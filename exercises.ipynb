Дан массив (10,2) координат, найти расстояние от каждой точки до каждой

```python
import scipy.spatial as ss
vector = np.random.random((10,2))
distance = ss.distance.cdist(vector, vector)

print(vector, distance)
```

Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

```python
vector = np.arange(1, 6)
zero = 3

vector1 = np.zeros(len(vector) + (len(vector)-1) * zero)
vector1[::zero+1] = vector
vector1
```

## Pandas - фильтрация данных

Считать данные по  [адресу](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 
Присвоить переменной chipo

```python
from pandas import read_csv
chipo = read_csv("chipotle.tsv", '\t')
```

Как много продуктов стоят более чем $10.00?

```python
chipo['item_price_float'] = chipo['item_price']
chipo.item_price_float = chipo.item_price_float.str[1:].astype(float)
chipo[chipo.item_price_float > 10] 
```

Какова цена каждого предмета? Напечатать столбцы item_name и item_price

```python
chipo.loc[:,['item_name','item_price']]
```

Отсортировать по имени предмета

```python
chipo.sort_values(by=['item_name'])
```

Сколько было заказано самого дорогого предмета?

```python
chipo.loc[chipo['item_name'] == chipo.sort_values(by=["item_price"]).tail(1)["item_name"].values[0]].groupby('item_name')['quantity'].sum()[0]
```

Сколько раз был заказан предмет "Veggie Salad Bowl"?

```python
len(chipo.loc[chipo['item_name'] == "Veggie Salad Bowl"].index)
```

Сколько раз люди заказали более чем 1 предмет "Canned Soda"?

```python
len(chipo.loc[(chipo['item_name'] == "Canned Soda") & (chipo["quantity"] > 1)].index)
```
